```python
import numpy as np
import pandas as pd
import torch as tr
# from tqdm import tqdm
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
# import seaborn as sns
```

```python
rnadist = pd.read_hdf('data/rnadist_f_all.h5')
secdist = pd.read_csv("data/SequenceIdentity_f_all.csv")
matrix=np.array(secdist.iloc[:,1:])
ids=secdist.columns[1:]
secdist = pd.DataFrame(matrix,index=ids,columns=ids)
```

```python
aux=np.array(secdist)
ones = np.ones([aux.shape[0],aux.shape[0]])
a_secdist = ones-aux
a_secdist = pd.DataFrame(a_secdist,index=ids,columns=ids)
a_secdist
```

```python
from sklearn.cluster import SpectralClustering
from sklearn.metrics.cluster import contingency_matrix

k=20
clustering = SpectralClustering(n_clusters=k,affinity='precomputed_nearest_neighbors').fit(rnadist)
rnadist['clusters']=clustering.labels_


clustering_sec = SpectralClustering(n_clusters=k,affinity='precomputed_nearest_neighbors').fit(a_secdist)
secdist['clusters']=clustering_sec.labels_

rnadist["clusters_sec"]=secdist['clusters']
```
